```python
import pandas as pd
import numpy as np
import scipy
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rc

rc('text', usetex=True)
rc('text.latex', preamble=r'\usepackage{cmbright}')
rc('font', **{'family': 'sans-serif', 'sans-serif': ['Helvetica']})

%matplotlib inline

# This enables SVG graphics inline. 
%config InlineBackend.figure_formats = {'png', 'retina'}

rc = {'lines.linewidth': 2, 
      'axes.labelsize': 18, 
      'axes.titlesize': 18, 
      'axes.facecolor': 'DFDFE5'}
sns.set_context('notebook', rc=rc)
sns.set_style("dark")

mpl.rcParams['xtick.labelsize'] = 16 
mpl.rcParams['ytick.labelsize'] = 16 
mpl.rcParams['legend.fontsize'] = 14
```

```python
res = pd.read_csv('../data/master_table.tsv', sep='\t')
```

```python
res['logMinCPM'] = res.MinCountsDetected.apply(np.log10)
g = sns.scatterplot('logMinCPM', 'log2FoldChange58', hue='Significance', alpha=0.1,
                    data=res[res.Significance != 'Not Sig'].rename(columns={'log2FoldChange_58': 'log2FoldChange58'}))

plt.ylabel('log2FC 58')
plt.xlabel('log10 CPM')
```
